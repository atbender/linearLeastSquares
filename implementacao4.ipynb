```python
import numpy as np
import matplotlib.pyplot as plt

# 1 Letra A
xi = np.array([1872, 1890, 1900, 1920, 1940, 1950, 1960, 1970, 1980, 1991, 1996])  
yi = np.array([9.9, 14.3, 17.4, 30.6, 41.2, 51.90, 70.2, 93.1, 119.0, 146.2, 157.1])  

xx = np.linspace(1870, 2020)  

plt.plot(xi,yi, 'ro')  
plt.show()
```

```python
# 1 Letra B

V = np.array([xi**2,xi**1,xi**0]).transpose()
a = ((np.linalg.inv((V.transpose()).dot(V))).dot(V.transpose())).dot(yi)

plt.plot(xi,yi,'ro',xx,np.polyval(a,xx),'b-')
plt.show()
```

```python
# 1 Letra D

xp = np.array([2000, 2005, 2014])
yp = np.array([169.8, 184.2, 202.7])

p = np.polyval(a, xp)

print('Valor estimados f(2000) = ' + str(np.polyval(a, 2000)))
print('Valor estimado f(2005) = ' + str(np.polyval(a, 2005)))
print('Valor estimado f(2014) = ' +  str(np.polyval(a, 2014)))
      
p = np.absolute(np.subtract(p, yp))
print('Erro Medio: ' + str(np.average(p)))

plt.plot(xi,yi, 'ro')
plt.plot(xp, yp, 'go', xx, np.polyval(a, xx), 'b-')
plt.show()
```

```python
# 1 Letra C

V = np.array([xi**1,xi**0]).transpose()
yi_log = np.log(yi)
a = ((np.linalg.inv((V.transpose()).dot(V))).dot(V.transpose())).dot(yi_log)

plt.plot(xi, yi, 'ro', xx, np.exp(np.polyval(a, xx)), 'b-')
plt.show()
```

```python
# 1 Letra D

p = np.exp(np.polyval(a, [2000, 2005, 2014]))

p = np.absolute(np.subtract(p, [169.8, 184.2, 202.7]))
print('Valor estimados f(2000) = ' + str(np.exp(np.polyval(a, 2000))))
print('Valor estimado f(2005) = ' + str(np.exp(np.polyval(a, 2005))))
print('Valor estimado f(2014) = ' + str(np.exp(np.polyval(a, 2014))))
print('Erro Medio: ' + str(np.average(p)))

plt.plot(xi,yi, 'ro')
plt.plot(xp, yp, 'go', xx, np.exp(np.polyval(a, xx)), 'b-')
plt.show()
```
